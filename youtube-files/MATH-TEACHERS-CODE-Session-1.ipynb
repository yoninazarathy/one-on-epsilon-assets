# Getting to business: The Hailstone Sequence

```julia
isEven(x) = x % 2 == 0 
next(x) = isEven(x) ? Int(x/2) : 3x+1
```

```julia
x = 7
println(x)
x = next(x)
println(x)
x = next(x)
println(x)
x = next(x)
println(x)
```

#### Loop attempt 1

```julia
x = 7
for _ in 1:12
    print(x," ")
    x = next(x)
end
```

#### A more suitable loop

```julia
function printHail(x)
    while x != 1
        print(x," ")
        x = next(x)
    end
    print(x," ")
end
```

```julia
printHail(323523559)
```

```julia
function hailArray(x)
    arrOut = [x]
    while x != 1
        x = next(x)
        push!(arrOut,x)
    end
    return arrOut
end
```

```julia
using PyPlot
start = 27
out = hailArray(start)
n = length(out)
plot(1:n,out)
```
